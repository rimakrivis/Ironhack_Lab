```python
import random
import time

def flip_coin():
    # Define possible outcomes
    outcomes = ["Heads", "Tails"]
    
    print("🪙 Flipping the coin...")
    # Add a small delay for suspense
    time.sleep(1)
    
    result = random.choice(outcomes)
    print(f"✨ It's: {result}!")
    return result

def start_game():
    stats = {"Heads": 0, "Tails": 0}
    
    print("--- Welcome to the Coin Flip Simulator ---")
    
    while True:
        result = flip_coin()
        stats[result] += 1
        
        print(f"Current Stats: Heads {stats['Heads']} | Tails {stats['Tails']}")
        
        play_again = input("\nFlip again? (y/n): ").lower()
        if play_again != 'y':
            print("Thanks for playing! Final score above.")
            break

if __name__ == "__main__":
    start_game()
```
